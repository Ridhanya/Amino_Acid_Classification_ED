```python
#dataframe to store data with their correlation values
df_corrsplit = pd.DataFrame()
```

```python
#make a copy of the dataframe - linear baseline normalized data
df_full_baseline = df_full
```

```python
#find correlation of each sequence with it's corresponding mean sequence
for i in df_full_baseline['labels'].unique():
  df_temp = df_full_baseline[df_full_baseline['labels']==i]


  corr = pd.Series(df_temp.iloc[:,:72].T.corrwith(df_temp.mean(axis=0),method='spearman'))
  df_temp['corrp'] = corr
  df_temp['labels'] = i

  df_corrsplit = pd.concat([df_corrsplit,df_temp],axis=0)
```

```python
#Filter 3 least correlated sequence and validate (following cells below)
df_corrank = pd.DataFrame()
for i in df_rank['labels'].unique():
  df_temp = df_rank[df_rank['labels']==i]
  df_temp_sort = df_temp.sort_values('corrp').iloc[3:,:]



  df_corrank = pd.concat([df_corrank,df_temp_sort],axis=0)
```
