```python
#import libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#normalization function
def normalizearr(arr):
  mean = np.mean(arr,axis=0)
  std = np.std(arr,axis=0)
  arrnorm= (arr-mean)/std
  return arrnorm
```

```python
#load may 3 data - linear baseline corrected
data3may = pd.read_csv('/content/drive/My Drive/ED_AA_AI/spring23_dataprocessed/23_may3/may3fullbaselinelinearnotnormalized.csv')
```

```python
#print shape
data3may.shape
```

```python
#apply normalization
for i in range(len(data3may)):
  data3may.iloc[i,:72] = normalizearr(np.array(data3may.iloc[i,:72]))
```
